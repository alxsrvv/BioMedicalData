```python
import seaborn as sns
%matplotlib inline
```

```python
planets = sns.load_dataset('planets')
```

```python
planets.head()
```

<h5>distplot - показує розподіл одновимірного набору спостережень</h5>

```python
sns.distplot(planets['year'])
```

```python
sns.distplot(planets['year'], kde = False, bins = 5)
```

<h5>jointplot – дозволяє співставити розподіл двох змінних та їх взаємозв’язок за допомогою діаграми розсіювання</h5>

```python
sns.jointplot(x = 'orbital_period', y = 'mass', data = planets)
```

```python
sns.jointplot(x = 'orbital_period', y = 'mass', data = planets,
             kind = 'kde')
```

<h5>pairplot – будує попарні взаємозв’язки між змінними в повному наборі даних</h5>

```python
sns.pairplot(planets)
```

```python
sns.pairplot(planets, hue = 'method')
```

<h5>barplot – дозволяють отримувати сукупні дані по кількісним даним в залежності від категоріальної змінної</h5>

```python
sns.barplot(x = 'number', y = 'mass', data = planets)
```

```python
import numpy as np
sns.barplot(x = 'number', y = 'mass', data=planets, estimator = np.median)
```

<h5>boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною</h5>

```python
sns.boxplot(x = 'number', y = 'mass', data = planets,
            palette = 'rainbow')
```

<h5>stripplot – представляє собою діаграму розсіювання, де одна за змінних категоріальна</h5>

```python
sns.stripplot(x = 'number', y = 'mass', data = planets)
```

<h5>Heatmap – графік взаємозв’язку (кореляції) між даними у вигляді матриці</h5>

```python
sns.heatmap(planets.corr(), annot = True)
```
